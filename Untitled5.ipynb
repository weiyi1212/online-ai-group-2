```python
import pandas as pd
import sklearn 
import seaborn as sns 
import numpy as np
import random
random.seed(5483)
```

```python
tf = pd.read_csv("tf_mini.csv")
```

```python
tf.describe()
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(tf, test_size=0.2)
```

```python
train
```

```python
test
```

```python
train = train.drop(("track_id"), axis=1)
train = train.drop(("release_year"), axis=1)
train = train.drop(("time_signature"), axis=1)
train = train.drop(("acoustic_vector_0"), axis=1)
train = train.drop(("acoustic_vector_1"), axis=1)
train = train.drop(("acoustic_vector_2"), axis=1)
train = train.drop(("mode"), axis=1)
```

```python
train
```

```python
test = test.drop(("track_id"), axis=1)
test = test.drop(("release_year"), axis=1)
test = test.drop(("time_signature"), axis=1)
test = test.drop(("acoustic_vector_0"), axis=1)
test = test.drop(("acoustic_vector_1"), axis=1)
test = test.drop(("acoustic_vector_2"), axis=1)
test = test.drop(("mode"), axis=1)
```

```python
test
```

```python
y = train.pop("danceability")
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression().fit(train, y)
```

```python
y_test = test.pop("danceability")
```

```python
predicted = model.predict(test)
```

```python
tf_test = pd.DataFrame(y_test)
```

```python
tf_test
```

```python
tf_test["predicted"] = predicted
```

```python
sns.lineplot("danceability","predicted", data = tf_test)
```

```python
tf_test
```

```python
tf_test.plot(x="danceability", y="predicted", kind= "line")
```
