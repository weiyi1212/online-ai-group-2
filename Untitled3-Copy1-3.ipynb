```python
import pandas as pd
import sklearn 
import seaborn as sns 
import numpy as np
import random
random.seed(5483)

```

```python
df = df.drop(["hour_of_day"], axis=1)
```

```python
from sklearn.model_selection import train_test_split

train, test = train_test_split(df, test_size=0.2)
```

```python
train
```

```python
test
```

```python
train = train.drop(("track_id_clean"), axis=1)
train = train.drop(("session_id"), axis=1)
train = train.drop(("date"), axis=1)
train = train.drop(("context_type"), axis=1)
train = train.drop(("hist_user_behavior_reason_start"), axis=1)
train = train.drop(("hist_user_behavior_reason_end"), axis=1)
```

```python
train
```

```python
test = test.drop(("track_id_clean"), axis=1)
test = test.drop(("session_id"), axis=1)
test = test.drop(("date"), axis=1)
test = test.drop(("context_type"), axis=1)
test = test.drop(("hist_user_behavior_reason_start"), axis=1)
test = test.drop(("hist_user_behavior_reason_end"), axis=1)
```

```python
test
```

```python
train = train.applymap(lambda x: 1 if x == True else x)
train = train.applymap(lambda x: 0 if x == False else x)
```

```python
train
```

```python
y = train.pop("not_skipped")
```

```python
from sklearn.linear_model import LinearRegression


model = LinearRegression().fit(train, y)
```

```python
test = test.applymap(lambda x: 1 if x == True else x)
test = test.applymap(lambda x: 0 if x == False else x)
```

```python
y_test = test.pop("not_skipped")
```

```python
predicted = model.predict(test)
```

```python
predicted
```

```python
df_test = pd.DataFrame(y_test)
```

```python
df_test["predicted"] = predicted
```
